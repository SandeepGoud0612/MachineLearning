```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
height = [150,160,165,185,250]
weight = [70, 80, 90, 100,110]
calories_burnt = [65, 75, 95, 99, 105]
```

```python
plt.figure(figsize=(10,5))
plt.plot(height, weight, 'r--', marker='*')
plt.xlabel('height')
plt.ylabel('weight')
plt.title('Relationship between height and weight')
```

```python
plt.figure(figsize=(15,5))
plt.plot(weight, marker='o')
plt.plot(calories_burnt, marker='*')
plt.title('relationship between weight and calories burnt')
plt.xlabel('persons')
plt.ylabel('weight & calories_burnt')
plt.xticks(ticks=[1,2,3,4,5], labels=['p1','p2','p3','p4','p5'])
plt.legend(labels=['weight', 'calories burnt'], loc='lower right')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,6), sharex=True, sharey=True)
ax[0].plot(calories_burnt, 'r--', marker='*')
ax[1].plot(weight, marker='o')
ax[0].set_title('Calories Burnt')
ax[1].set_title('Weight')
ax[0].set_xticks(ticks=[1,2,3,4,5])
ax[1].set_xticks(ticks=[1,2,3,4,5])
ax[0].set_xticklabels(labels=['p1','p2','p3','p4'])
ax[1].set_xticklabels(labels=['p1','p2','p3','p4'])
```

```python
df = pd.read_csv('../dataset/csv/bigmart_data.csv')
df.head()
```

```python
df1 = df.groupby(['Item_Type']).Item_MRP.mean()[:6]
df1
```

```python
x = df1.index.tolist()
y = df1.values.tolist()
plt.figure(figsize=(15,5))
plt.plot(x, y, 'r--', marker='*')
plt.xlabel('Item Type')
plt.ylabel('Item MRP Mean')
plt.title('Mean Price for each Item type')
```

```python
df2 = df.groupby(['Item_Type'])
df2.Item_MRP.mean()
```

```python
df3 = df.groupby(['Outlet_Size']).Item_Outlet_Sales.mean()
df3
```

```python
plt.figure(figsize=(15,8))
x = df3.index.tolist()
y = df3.values.tolist()
plt.bar(x, y, color=['red', 'green', 'yellow'])
plt.xlabel('Outlet Size')
plt.ylabel('Item Outlet Sales Mean')
plt.title('Mean of Item Outlet Sales of each Outlet Size')
```

```python
plt.figure(figsize=(20,8))
plt.xlabel('Item MRP')
plt.ylabel('Frequency')
plt.title('Item MRP Distribution')
plt.hist(df['Item_MRP'], bins=150, color='green')
```

```python
plt.figure(figsize=(10,8))
df4 = df.Item_Outlet_Sales
red_diamond = dict(markerfacecolor='r', marker='D')
plt.title('Item Sales Distribution')
plt.boxplot(df4.values, labels=['Item Sales'], flierprops=red_diamond)
```

```python
df5 = df.Item_MRP
plt.xlabel('Item MRP')
plt.violinplot(df5.values)
```

```python
plt.xlabel('Item Weight')
plt.ylabel('Item Outlet Sales')
plt.scatter(df['Item_Weight'][:100], df['Item_Outlet_Sales'][:100])
```

```python
plt.xlabel('Item Weight')
plt.ylabel('Item Outlet Sales')
plt.scatter(df['Item_Weight'][:100], df['Item_Outlet_Sales'][:100], s=df['Item_Visibility'][:100]*1000, c='red')
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set()

import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize']=(10,10)
```

```python
data = pd.read_csv('../dataset/csv/bigmart_data.csv')
data.shape
```

```python
data.isna().sum()
```

```python
data = data.dropna(how='any')
data.shape
data['Visibility_Scaled'] = data['Item_Visibility'] * 100
data.head()
```

```python
sns.lineplot(x='Item_Weight', y='Item_MRP', data=data[:50])
```
