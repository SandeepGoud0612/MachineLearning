```python
data = pd.read_csv('../dataset/csv/bigmart_data.csv')
data = data.dropna(how='any')
data['Visibility_Scaled'] = data['Item_Visibility'] * 100
data.head()
```

```python
sns.lineplot(x='Item_Weight', y='Item_MRP', data=data[:50])
```
