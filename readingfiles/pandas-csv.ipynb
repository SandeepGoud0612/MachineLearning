```python
from glob import glob # Use the glob library to list the files in the directory
```

```python
for directory in glob('../dataset/csv/multi-directory/*'):
    print(directory)
```

```python
for directory in glob.glob('dataset/multi-directory/*'):
    for files in glob.glob(directory + '/*'):
        print(files)
```

```python
data_frame_list = []
for directory in glob.glob('dataset/multi-directory/*'):
    for files in glob.glob(directory + '/*'):
        data_frame_list.append(pd.read_csv(files))
```

```python
data_frame_list
```

```python
final_data = pd.concat(data_frame_list)
```

```python
final_data
```

```python
final_data.shape
```

```python
final_data.head()
```

```python
final_data.to_csv('dataset/merged_big_mart_sales.csv', index=False)
```
